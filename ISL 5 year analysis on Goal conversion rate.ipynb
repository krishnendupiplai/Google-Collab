```python
#! pip install streamlit -q

from plotly.offline import iplot
import plotly.graph_objs as go
import plotly.express as px

#import streamlit as st
import pandas as pd
import requests
import numpy as np

'''url6= 'https://fbref.com/en/comps/82/2022-2023/shooting/2022-2023-Indian-Super-League-Stats'
response6 = requests.get(url6).text.replace('<!--', '').replace('-->', '')
Playergoal2022 = pd.read_html(response6, header=1)[2]
#Playergoal2021.columns=Playergoal2021.columns.droplevel()
Playergoal2022=Playergoal2022.drop(columns=["Matches","Age"])
Playergoal2022=Playergoal2022.drop(Playergoal2022[Playergoal2022["Nation"]=="Nation"].index)
#U23shooting["AGE"]=2021-U23shooting["Born"]
Playergoal2022["Born"]=pd.to_numeric(Playergoal2022["Born"])
Playergoal2022["Season"]=2022
Playergoal2022["Age"]=Playergoal2022["Season"]-Playergoal2022["Born"]'''



url1= 'https://fbref.com/en/comps/82/shooting/Indian-Super-League-Stats'
response1 = requests.get(url1).text.replace('<!--', '').replace('-->', '')
Playergoal2021 = pd.read_html(response1, header=1)[2]
#Playergoal2021.columns=Playergoal2021.columns.droplevel()
Playergoal2021=Playergoal2021.drop(columns=["Matches","Age"])
Playergoal2021=Playergoal2021.drop(Playergoal2021[Playergoal2021["Nation"]=="Nation"].index)
#U23shooting["AGE"]=2021-U23shooting["Born"]
Playergoal2021["Born"]=pd.to_numeric(Playergoal2021["Born"])
Playergoal2021["Season"]=2021
Playergoal2021["Age"]=Playergoal2021["Season"]-Playergoal2021["Born"]


url2= 'https://fbref.com/en/comps/82/2020-2021/shooting/2020-2021-Indian-Super-League-Stats'
response2 = requests.get(url2).text.replace('<!--', '').replace('-->', '')
Playergoal2020 = pd.read_html(response2, header=1)[2]
#Playergoal2020.columns=Playergoal2020.columns.droplevel()
Playergoal2020=Playergoal2020.drop(columns=["Matches","Age"])
Playergoal2020=Playergoal2020.drop(Playergoal2020[Playergoal2020["Nation"]=="Nation"].index)
Playergoal2020["Born"]=pd.to_numeric(Playergoal2020["Born"])
#U23shooting["AGE"]=2021-U23shooting["Born"]
Playergoal2020["Season"]=2020
Playergoal2020["Age"]=Playergoal2020["Season"]-Playergoal2020["Born"]


url3= 'https://fbref.com/en/comps/82/2020-2021/shooting/2020-2021-Indian-Super-League-Stats'
response3 = requests.get(url3).text.replace('<!--', '').replace('-->', '')
Playergoal2019 = pd.read_html(response3, header=1)[2]



#Playergoal2019.columns=Playergoal2019.columns.droplevel()
Playergoal2019=Playergoal2019.drop(columns=["Matches","Age"])
Playergoal2019=Playergoal2019.drop(Playergoal2019[Playergoal2019["Nation"]=="Nation"].index)
#U23shooting["AGE"]=2021-U23shooting["Born"]
Playergoal2019["Season"]=2019
Playergoal2019["Born"]=pd.to_numeric(Playergoal2019["Born"])
Playergoal2019["Age"]=Playergoal2019["Season"]-Playergoal2019["Born"]


url4= 'https://fbref.com/en/comps/82/2020-2021/shooting/2020-2021-Indian-Super-League-Stats'
response4 = requests.get(url4).text.replace('<!--', '').replace('-->', '')
Playergoal2018 = pd.read_html(response4, header=1)[2]



#Playergoal2018.columns=Playergoal2018.columns.droplevel()
Playergoal2018=Playergoal2018.drop(columns=["Matches","Age"])
Playergoal2018=Playergoal2018.drop(Playergoal2018[Playergoal2018["Nation"]=="Nation"].index)
#U23shooting["AGE"]=2021-U23shooting["Born"]
Playergoal2018["Season"]=2018
Playergoal2018["Born"]=pd.to_numeric(Playergoal2018["Born"])
Playergoal2018["Age"]=Playergoal2018["Season"]-Playergoal2018["Born"]


url5= 'https://fbref.com/en/comps/82/2017-2018/shooting/2017-2018-Indian-Super-League-Stats'
response5 = requests.get(url5).text.replace('<!--', '').replace('-->', '')
Playergoal2017 = pd.read_html(response5, header=1)[2]



#Playergoal2018.columns=Playergoal2018.columns.droplevel()
Playergoal2017=Playergoal2017.drop(columns=["Matches","Age"])
Playergoal2017=Playergoal2017.drop(Playergoal2017[Playergoal2017["Nation"]=="Nation"].index)
#U23shooting["AGE"]=2021-U23shooting["Born"]
Playergoal2017["Season"]=2017
Playergoal2017["Born"]=pd.to_numeric(Playergoal2017["Born"])
Playergoal2017["Age"]=Playergoal2017["Season"]-Playergoal2017["Born"]



compareShoot=pd.concat([Playergoal2021,Playergoal2020,Playergoal2019,Playergoal2018,Playergoal2017])
compareShoot[['Born', '90s','Sh','Gls','SoT','SoT%','SoT/90','G/Sh','G/SoT']] = compareShoot[['Born', '90s','Sh','SoT','Gls','SoT%','SoT/90','G/Sh','G/SoT']].astype(float)
compareShoot["Age group"]=np.where(compareShoot["Age"]<24,"U23","Senior")
compareShoot["Quantity - Shot convert rate"]=compareShoot["G/Sh"]*100
compareShoot["Quality - Shot convert rate"]=compareShoot["G/SoT"]*100
Senior_U23_compare=pd.DataFrame(compareShoot[compareShoot["Nation"]=="in IND"].groupby(["Season","Age group"])["Quality - Shot convert rate"].mean()).reset_index().rename(columns={"Quality - Shot convert rate":"Avg. Quality - Shot convert rate"})

from plotly.offline import iplot
import plotly.graph_objs as go
import plotly.express as px

#data_teacoffee = px.data.gapminder()
fig = px.line(Senior_U23_compare, y='Avg. Quality - Shot convert rate', x='Season',color="Age group",title="ISL 5-year Analysis<br>Indian Avg. Quality - Shot convert rate<br>Sourced from fbref by Footilytics",
              width=644,height=444,labels=[None,"2017",None,"2018",None,"2019",None,"2020",None,"2021"],markers=True)
fig
```
