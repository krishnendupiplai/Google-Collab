```python
import pandas as pd; import numpy as np; import matplotlib.pyplot as plt

data = pd.read_csv('https://raw.githubusercontent.com/holoviz/panel/master/examples/assets/occupancy.csv')
data['date'] = data.date.astype('datetime64[ns]')
data = data.set_index('date')

data.tail()
```

```python
from matplotlib.figure import Figure
from matplotlib.backends.backend_agg import FigureCanvas

%matplotlib inline

def mpl_plot(avg, highlight):
    fig = Figure()
    FigureCanvas(fig) # not needed in mpl >= 3.1
    ax = fig.add_subplot()
    avg.plot(ax=ax)
    if len(highlight): highlight.plot(style='o', ax=ax)
    return fig

def find_outliers(variable='Temperature', window=30, sigma=10, view_fn=mpl_plot):
    avg = data[variable].rolling(window=window).mean()
    residual = data[variable] - avg
    std = residual.rolling(window=window).std()
    outliers = (np.abs(residual) > std * sigma)
    return view_fn(avg, avg[outliers])
```

```python
find_outliers(variable='Temperature', window=20, sigma=10)
```

```python
YearlyAvgReturn=All_fund.groupby("Type").mean("Mean Return over tenurity(excl YTD)").loc[:,:"2011"]
YearlyAvgReturn
```

```python
YearlyAvgReturn=YearlyAvgReturn[['2011', '2012','2013','2014','2015','2016','2017','2018','2019','2021']]
YearlyAvgReturn
```

```python
plt.style.use("default")

YearlyAvgReturn.T.plot(kind="line")

plt.legend(loc='center left', bbox_to_anchor=(0.5, 0.5))
```

```python
for_Rel=YearlyAvgReturn.reset_index().melt(id_vars="Type").rename(columns={"variable":"Calender Year","value":"Return%"})

for_Rel
```

```python
plt.style.use("seaborn")
sns.relplot(data=for_Rel,x="Calender Year",y="Return%",kind="line",style='Type',hue="Type")

plt.title("Mutual fund historical return analysis Analysis\nReturn generated in a calenar year\nBuilt on python and sourced from moneycontrol.com by NowSufFermyWork",color="b")

```

```python
All_fund.loc[:,["Type","YTD"]].groupby("Type").mean().sort_values(by="YTD",ascending=True).plot(kind="barh",title="Mutual fund 2022 YTD return analysis\nBuilt on python and sourced from moneycontrol.com by NowSufFermyWork",color="b")
```

```python
plt.style.use("default")
sns.relplot(data=for_Rel,
            x="Calender Year",col="Type",y="Return%",kind="line",height=5,col_wrap=3,style='Type',facet_kws={'sharey': False, 'sharex': False}).fig.subplots_adjust(top=0.9)

plt.suptitle("Mutual fund historical return analysis Analysis\nReturn generated in a calenar year\nBuilt on python and sourced from moneycontrol.com by NowSufFermyWork",color="b")

```
